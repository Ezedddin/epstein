```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('to_db/epstein.db')
cursor = conn.cursor()

# Count flights per year based on flight_date
rows = cursor.execute(
    '''
    SELECT SUBSTR(flight_date, 1, 4) AS year, COUNT(*) AS flight_count
    FROM flights
    WHERE flight_date IS NOT NULL
      AND TRIM(flight_date) != ''
    GROUP BY year
    ORDER BY year
    '''
).fetchall()

conn.close()

# Main list with a sublist per year:
# [year, flight_count, growth_percentage]
yearly_growth_list = []

previous_count = None
for year, count in rows:
    if previous_count is None or previous_count == 0:
        growth_pct = None  # no growth value for first year
    else:
        growth_pct = ((count - previous_count) / previous_count) * 100

    yearly_growth_list.append([int(year), int(count), growth_pct])
    previous_count = count

print('Yearly growth list:')
for item in yearly_growth_list:
    year, count, growth = item
    growth_text = 'N/A' if growth is None else f'{growth:.2f}%'
    print(f'Year: {year} | Flights: {count} | Growth: {growth_text}')

yearly_growth_list
```
