```python
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('epstein.db')
cursor = conn.cursor()

# Count flights per year based on flight_date
rows = cursor.execute(
    '''
    SELECT SUBSTR(flight_date, 1, 4) AS year, COUNT(*) AS flight_count
    FROM flights
    WHERE flight_date IS NOT NULL
      AND TRIM(flight_date) != ''
    GROUP BY year
    ORDER BY year
    '''
).fetchall()

conn.close()

# Main list with a sublist per year:
# [year, flight_count, growth_percentage]
yearly_growth_list = []

previous_count = None
for year, count in rows:
    if previous_count is None or previous_count == 0:
        growth_pct = None  # no growth value for first year
    else:
        growth_pct = ((count - previous_count) / previous_count) * 100

    yearly_growth_list.append([int(year), int(count), growth_pct])
    previous_count = count

print('Yearly growth list:')
for item in yearly_growth_list:
    year, count, growth = item
    growth_text = 'N/A' if growth is None else f'{growth:.2f}%'
    print(f'Year: {year} | Flights: {count} | Growth: {growth_text}')

yearly_growth_list
```

```python
# Visualize year-over-year growth rate
years = [item[0] for item in yearly_growth_list if item[2] is not None]
growth_rates = [item[2] for item in yearly_growth_list if item[2] is not None]

fig, ax = plt.subplots(figsize=(12, 5))
colors = ["#2ca02c" if g >= 0 else "#d62728" for g in growth_rates]

ax.bar(years, growth_rates, color=colors, edgecolor="white")
ax.axhline(0, color="black", linewidth=1)
ax.set_title("Year-over-year flight growth rate")
ax.set_xlabel("Year")
ax.set_ylabel("Growth rate (%)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Average passenger_count per weather type (departure weather)
conn = sqlite3.connect('epstein.db')
cursor = conn.cursor()

weather_avg_rows = cursor.execute(
    '''
    SELECT
        wc.code AS weather_type,
        ROUND(AVG(f.passenger_count), 2) AS avg_passenger_count,
        COUNT(f.id) AS flights_n
    FROM weather_conditions wc
    LEFT JOIN flights f ON f.departure_weather_id = wc.id
    GROUP BY wc.code
    ORDER BY wc.code
    '''
).fetchall()

conn.close()

print('Average passenger_count by weather type:')
for weather_type, avg_count, flights_n in weather_avg_rows:
    print(f'{weather_type}: avg={avg_count}, flights={flights_n}')

labels = [row[0] for row in weather_avg_rows]
avg_values = [float(row[1]) if row[1] is not None else 0.0 for row in weather_avg_rows]

fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(labels, avg_values, color=['#f1c40f', '#3498db', '#95a5a6'], edgecolor='white')

ax.set_title('Average passenger count by weather type')
ax.set_xlabel('Weather type')
ax.set_ylabel('Average passenger_count')

for bar, val in zip(bars, avg_values):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{val:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
